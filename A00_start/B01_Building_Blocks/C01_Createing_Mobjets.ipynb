```python
from manim import *
```

```python
%%manim --format gif -ql -v Warning CreatingMobjets
class CreatingMobjets(Scene):
    def construct(self):
        circle = Circle()
        self.add(circle)
        self.wait(2)
        self.remove(circle)
        self.wait(2)
```

```python
%%manim --format gif -ql -v Warning Shapes

class Shapes(Scene):
    def construct(self):
        circle = Circle()
        square = Square()
        triangle = Triangle()

        circle.shift(LEFT)
        square.shift(UP)
        triangle.shift(RIGHT)

        self.add(circle, square, triangle)
        self.wait(1)
```

```python
%%manim --format gif -ql -v Warning MobjectPlacement

class MobjectPlacement(Scene):
    def construct(self):
        circle = Circle()
        square = Square()
        triangle = Triangle().set_fill(opacity=1.0)

        circle.move_to(LEFT * 2)
        square.next_to(circle, LEFT)
        triangle.align_to(circle, LEFT)

        # 체인룰도 적용할 수 있음. Mobject 는 수정된 Mobjetc를 반환하기 때문. 
        self.add(circle, square, triangle)
        self.wait(1)
```

```python
%%manim --format gif -ql -v Warning MobjectStyling

class MobjectStyling(Scene):
    def construct(self):
        circle = Circle().shift(LEFT).set_stroke(color=GREEN, width=20)
        square = Square().shift(UP).set_fill(YELLOW, opacity=1.0)
        triangle = Triangle().shift(RIGHT).set_fill(PINK, opacity=1.0).set_stroke(color=PINK)

        self.add(circle, square, triangle)
        self.wait(1)
```

```python
%%manim --format gif -ql -v Warning MobjectZOrder

class MobjectZOrder(Scene):
    def construct(self):
        circle = Circle().shift(LEFT).set_stroke(color=GREEN, width=20)
        square = Square().shift(UP).set_fill(YELLOW, opacity=1.0)
        triangle = Triangle().shift(RIGHT).set_fill(PINK, opacity=1.0).set_stroke(color=PINK)

        self.add(triangle, circle, square)
        self.wait(1)
```
